3.check whether a year is leap year

```python
year=int(input("enter a year\n"))
if(year%4==0) or (year%4==0 and year%100!=0):
  print("leap year")
else:
         print("not a leap year")
```

6.check whether a number is prime

```python
num = int(input("enter a number\n"))
flag=0
for i in range(2,num):
    if num%i==0:
       flag=1
       break

if flag == 0 and num>1:
          print("prime number ")
else:
          print("not a prime number")
```

7.find factorial of a number

```python
n=int(input("enter a number\n"))
fact=1
for i in range(1,n+1):
  fact=fact*i
print("factorial=",fact)
```

8.reverse a number

```python
num=int(input("enter a number\n"))
rev=0
while num>0:
  digit=num%10
  rev=rev*10+digit
  num=num//10
print("reversed number=",rev)
```

9.check whether a number is palindrome

```python
num=int(input("enter a number\n"))
temp=num
rev=0
while num>0:
  rev=rev*10+(num%10)
  num=num//10
if temp==rev:
  print("palidrome")
else:
  print("not a palidrome")
```

3.student marks analysis using tuple

```python
marks = tuple(map(int,input("enter marks\n").split()))
flag=0
for m in marks:
  if m<40:
    flag=1
    break


if flag ==0:
    print("student passed all subjects")
else:
  print("student failed")
```
